# Análise Exploratória: Migração de Usuários do Windows para Linux

## Tema
Estudo exploratório sobre possíveis fatores associados à migração de usuários do Windows para Linux, com foco em:
- Consumo de RAM e recursos do sistema
- Sobrecarga do sistema operacional Windows
- Percepção de flexibilidade


## Introdução

Este notebook realiza uma análise exploratória simulada de dados relacionados à migração de usuários do Windows para Linux.
O objetivo é investigar padrões associados ao consumo de memória (RAM), uso de disco (GB) e percepção de desempenho do sistema operacional.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```
# Dataset simulado
np.random.seed(42)

n = 200

data = pd.DataFrame({
    "ram_gb_usado_windows": np.random.normal(8, 2.5, n).clip(2, 32),
    "uso_disco_gb_windows": np.random.normal(120, 40, n).clip(30, 500),
    "travamentos_mes": np.random.poisson(3, n),
    "tempo_resposta_ms": np.random.normal(180, 60, n).clip(50, 500),
    "migracao_para_linux": np.random.choice([0, 1], size=n, p=[0.6, 0.4])
})

data.head()
```

## Estatísticas descritivas

```
data.describe()
```

## Relação entre uso de RAM e migração

```
plt.figure()
plt.scatter(data["ram_gb_usado_windows"], data["migracao_para_linux"])
plt.xlabel("RAM usada no Windows (GB)")
plt.ylabel("Migração para Linux (0 = não, 1 = sim)")
plt.title("Relação entre consumo de RAM e migração")
plt.show()
```

## Travamentos mensais vs migração

```
plt.figure()
plt.scatter(data["travamentos_mes"], data["migracao_para_linux"])
plt.xlabel("Travamentos por mês")
plt.ylabel("Migração para Linux")
plt.title("Instabilidade e migração de sistema")
plt.show()
```

## Correlação entre variáveis

```
data.corr(numeric_only=True)
```

## Conclusão

A análise simulada sugere que aumentos no uso de RAM, maior número de travamentos e maior latência podem estar associados a maior probabilidade de migração para sistemas mais leves e flexíveis.

> Observação: Este conjunto de dados é simulado para fins acadêmicos e não representa medições reais.
