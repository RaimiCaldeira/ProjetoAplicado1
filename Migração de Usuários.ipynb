# Análise Exploratória: Usuários de Windows e seus concorrentes

## Perda e clientes 
Estudo exploratório sobre possíveis fatores associados à migração de usuários do Windows para Linux, com foco em:
- Consumo de RAM e recursos do sistema
- Sobrecarga do sistema operacional Windows
- Percepção de flexibilidade


## Introdução

Este notebook realiza uma análise exploratória com dados reais de usuários por sistema operacional Windows e concorrentes. Dados usados pela plataforma Statista.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
# Número de usuários por sistema operacional
pd.set_option('display.max_columns', None)
labels = ['Windows', 'Macintosh', 'iOS', 'Android', 'Linux', 'Outros']
sizes = [72.9, 16.7, 3.3, 2.2, 1.8, 3.2]

plt.figure(figsize=(8,8))

plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%'
)

plt.title('Participação de Sistemas Operacionais')

plt.show()
```

## Estatísticas descritivas

Também demonstra o número crescente de funcionários da empresa Microsoft, que em outras palavras, a empresa só cresce e tende a crescer no projetamento do futuro, porém, até quando vale apena perder clientes e ignorar o fator de alto consumo de Gb's e RAM para outras empresas? Até quando essa perda de dinheiro vai se escorrer?

```python
anos = [2005, 2007, 2009, 2011, 2013, 2015, 2017, 2019, 2020, 2021, 2022]
funcionarios = [61, 49, 93, 90, 99, 118, 124, 144, 181, 181, 221]

plt.figure(figsize=(12,6))

plt.plot(
    anos,
    funcionarios,
    color='#0B5ED7',
    linewidth=3,
    marker='o'
)

for x, y in zip(anos, funcionarios):
    plt.text(
        x,
        y + 4,
        str(y),
        ha='center',
        fontsize=10,
        fontweight='bold'
    )

plt.title('Nº de funcionários (milhares)', fontsize=18, fontweight='bold')

plt.xlabel('Ano')
plt.ylabel('Funcionários (milhares)')

plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.xticks(anos, rotation=45)

plt.tight_layout()

plt.show()
```

Apesar de a Microsoft continuar registrando forte crescimento financeiro, impulsionado principalmente por serviços como o Microsoft Microsoft Azure e o Microsoft 365 (antigo Office), a empresa ainda enfrenta preocupações relacionadas ao futuro do Windows e à experiência dos usuários com o sistema operacional. O Windows é frequentemente criticado pelo alto consumo de memória RAM, uso elevado de armazenamento e limitações de flexibilidade e personalização, fatores que podem afetar o desempenho mesmo em computadores mais modernos. Em contrapartida, sistemas Linux são vistos por muitos usuários como alternativas mais leves, estáveis e eficientes no gerenciamento de recursos do hardware.

Embora outras divisões da Microsoft, como computação em nuvem e softwares corporativos, gerem receitas bilionárias e sustentem grande parte do crescimento da empresa, uma perda contínua de usuários do Windows ainda pode representar impactos financeiros relevantes. Isso ocorre porque o sistema operacional permanece como um dos pilares do ecossistema da Microsoft, influenciando vendas de licenças, integração com serviços da empresa e fidelização de usuários. Dessa forma, mesmo com o sucesso de áreas como Azure e Microsoft 365, uma migração significativa de usuários para sistemas alternativos pode gerar prejuízos de bilhões de dólares no longo prazo e reduzir a influência da Microsoft no mercado de sistemas operacionais para computadores pessoais.

## Conclusão

Apesar do forte crescimento da Microsoft com serviços como Azure e Microsoft 365, o Windows ainda enfrenta críticas pelo alto consumo de RAM, uso elevado de recursos e falta de flexibilidade. Esses problemas têm incentivado parte dos usuários a migrar para sistemas mais leves e eficientes, como o Linux. Mesmo sendo uma empresa bilionária, a perda gradual de usuários do Windows pode gerar impactos financeiros relevantes e reduzir a força da Microsoft no mercado de sistemas operacionais.
